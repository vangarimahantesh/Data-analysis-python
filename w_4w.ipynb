```python
import pandas as pd
import matplotlib.pyplot as plt

# File paths
file1 = "/content/daily-minimum-temperatures-in-me.csv"
file2 = "/content/Electric_Production.csv"
file3 = "/content/monthly-beer-production-in-austr.csv"
file4 = "/content/sales-of-shampoo-over-a-three-ye.csv"

# Load datasets
df1 = pd.read_csv(file1)
df2 = pd.read_csv(file2)
df3 = pd.read_csv(file3)
df4 = pd.read_csv(file4)

# Show basic info
print("Dataset 1 - Daily Min Temperatures")
print(df1.head(), "\n", df1.info(), "\n")

print("Dataset 2 - Electric Production")
print(df2.head(), "\n", df2.info(), "\n")

print("Dataset 3 - Beer Production")
print(df3.head(), "\n", df3.info(), "\n")

print("Dataset 4 - Shampoo Sales")
print(df4.head(), "\n", df4.info(), "\n")

# Plotting each dataset
plt.figure(figsize=(12,6))
plt.plot(df1.iloc[:,0], df1.iloc[:,1])
plt.title("Daily Minimum Temperatures")
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12,6))
plt.plot(df2.iloc[:,0], df2.iloc[:,1])
plt.title("Electric Production")
plt.xlabel("Date")
plt.ylabel("Production")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12,6))
plt.plot(df3.iloc[:,0], df3.iloc[:,1])
plt.title("Monthly Beer Production")
plt.xlabel("Date")
plt.ylabel("Production")
plt.xticks(rotation=45)
plt.show()

plt.figure(figsize=(12,6))
plt.plot(df4.iloc[:,0], df4.iloc[:,1])
plt.title("Shampoo Sales Over 3 Years")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```
