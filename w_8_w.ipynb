```python
import math
from scipy.stats import binom
import matplotlib.pyplot as plt

# Parameters
n = 20   # number of questions
p = 0.25 # probability of correct guess

# Distribution characteristics
mean = n * p
variance = n * p * (1 - p)
std_dev = math.sqrt(variance)
mode = math.floor((n + 1) * p)

print("=== Distribution Characteristics ===")
print(f"Mean: {mean}")
print(f"Variance: {variance}")
print(f"Standard Deviation: {std_dev:.4f}")
print(f"Most likely number of correct answers (mode): {mode}")

# Exact probability: at least 10 correct
prob_at_least_10 = 1 - binom.cdf(9, n, p)
print("\n=== Exact Probability Question ===")
print(f"P(X >= 10): {prob_at_least_10:.6f} ({prob_at_least_10*100:.2f}%)")

# Performance evaluation
prob_less_than_5 = binom.cdf(4, n, p)
prob_between_5_and_8 = binom.cdf(8, n, p) - binom.cdf(4, n, p)

print("\n=== Performance Evaluation ===")
print(f"P(X < 5): {prob_less_than_5:.6f} ({prob_less_than_5*100:.2f}%)")
print(f"P(5 <= X <= 8): {prob_between_5_and_8:.6f} ({prob_between_5_and_8*100:.2f}%)")

# Full probability distribution table
print("\n=== Full Probability Distribution (PMF) ===")
for k in range(n + 1):
    prob = binom.pmf(k, n, p)
    print(f"{k:2d} correct -> P(X={k}) = {prob:.6f} ({prob*100:.3f}%)")

# Plot the probability distribution
x = range(n + 1)
pmf = binom.pmf(x, n, p)

plt.bar(x, pmf, color="skyblue", edgecolor="black")
plt.title("Binomial Distribution: n=20, p=0.25")
plt.xlabel("Number of Correct Answers")
plt.ylabel("Probability")
plt.show()
```
