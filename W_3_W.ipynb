```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Simulate phishing email dataset
np.random.seed(42)
n = 200
data = {
    "account": np.random.poisson(2, n),
    "login": np.random.poisson(3, n),
    "spam": np.random.choice([0, 1], n)  # 0 = non-spam, 1 = spam
}
df_email = pd.DataFrame(data)

# Scatter plot for phishing detection
plt.figure(figsize=(7,5))
spam = df_email[df_email["spam"] == 1]
non_spam = df_email[df_email["spam"] == 0]
plt.scatter(non_spam["account"], non_spam["login"], color="blue", marker="o", label="Non-Spam")
plt.scatter(spam["account"], spam["login"], color="red", marker="*", label="Spam")
plt.title("Account & Login Word Frequency in Spam vs Non-Spam Emails")
plt.xlabel("Count of 'account'")
plt.ylabel("Count of 'login'")
plt.legend()
plt.show()

# Scatter plot for hours studied vs test scores
hours = np.random.randint(1, 10, 30)
scores = hours * 10 + np.random.randint(-5, 6, 30)
plt.figure(figsize=(7,5))
plt.scatter(hours, scores, color="green", marker="s", label="Students")
plt.scatter(hours[scores > 80], scores[scores > 80], color="red", marker="o", label="High Scores")
plt.title("Hours Studied vs Test Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Test Score")
plt.legend()
plt.show()
```
