```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
np.random.seed(42)
blood_pressure_data = np.random.normal(loc=130, scale=15, size=2000)
bp_df = pd.DataFrame(blood_pressure_data, columns=['Systolic_BP'])
plt.figure(figsize=(8, 6))
plt.hist(bp_df['Systolic_BP'], bins=30, color='skyblue', edgecolor='black')
plt.title('Histogram of Full Dataset (Systolic Blood Pressure)', fontsize=14)
plt.xlabel('Systolic Blood Pressure (mmHg)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
mean_bp = np.mean(blood_pressure_data)
median_bp = np.median(blood_pressure_data)
variance_bp = np.var(blood_pressure_data)
std_dev_bp = np.std(blood_pressure_data)
```

```python
print("Full Dataset Statistics:")
print(f"Mean: {mean_bp:.2f} mmHg")
print(f"Median: {median_bp:.2f} mmHg")
print(f"Variance: {variance_bp:.2f} mmHg^2")
print(f"Standard Deviation: {std_dev_bp:.2f} mmHg")
def perform_sampling(sample_size, n_samples=1000):
    sample_means = []
    for _ in range(n_samples):
        sample = np.random.choice(blood_pressure_data, size=sample_size, replace=False)
        sample_means.append(np.mean(sample))

    return sample_means
sample_sizes = [25, 50, 100]
```

```python
plt.tight_layout()
plt.show()
# Plot histograms of the sample means for each sample size
plt.figure(figsize=(14, 10))

for i, sample_size in enumerate(sample_sizes, 1):
    sample_means = perform_sampling(sample_size)
    plt.subplot(2, 2, i)
    plt.hist(sample_means, bins=30, color='lightcoral', edgecolor='black')
    plt.title(f"Sample Means for Sample Size = {sample_size}", fontsize=14)
    plt.xlabel('Mean Systolic BP (mmHg)', fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
```

```python
# 4. Adaptation - Allow the user to set custom sample sizes (e.g., 40, 75, 150)
custom_sample_size = 40  # Change this to any other number (e.g., 75, 150) for testing
custom_sample_means = perform_sampling(custom_sample_size)

# Plot sample means for the custom sample size
plt.figure(figsize=(8, 6))
plt.hist(custom_sample_means, bins=30, color='lightgreen', edgecolor='black')
plt.title(f"Sample Means for Custom Sample Size = {custom_sample_size}", fontsize=14)
plt.xlabel('Mean Systolic BP (mmHg)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.show()
```

```python
# 5. Line Plot to Show How Standard Error Decreases with Increasing Sample Size
sample_sizes_range = range(10, 201, 10)  # Sample sizes from 10 to 200 in steps of 10
std_errors = []

for size in sample_sizes_range:
    sample_means = perform_sampling(size)
    std_errors.append(np.std(sample_means))

plt.figure(figsize=(8, 6))
plt.plot(sample_sizes_range, std_errors, marker='o', color='blue')
plt.title('Standard Error vs. Sample Size', fontsize=14)
plt.xlabel('Sample Size', fontsize=12)
plt.ylabel('Standard Error', fontsize=12)
plt.grid(True)
plt.show()
```
